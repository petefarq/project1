```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

```python
file = "Resources/AQ_health_econ.csv"
file_df = pd.read_csv(file)
file_df
```

```python
file_df[["City", "PM2.5 2014", "Poverty Rate 2009-2013"]].head()
```

```python
pollution = file_df["PM2.5 2014"]
poverty_rate  = file_df["Poverty Rate 2009-2013"]

plt.scatter(poverty_rate, pollution, marker="o", facecolors="red", edgecolors="black",
            alpha=0.75)

plt.legend(loc="best", fontsize="small", fancybox=True)
plt.title("Poverty Rate vs. PM2.5 2014")

plt.xlabel("Poverty Rate 2009-2013")
plt.ylabel("PM2.5 2014 g/m3 (micrograms per cubic meter)")

(slope, intercept, r_value, p_value, std_err) = stats.linregress(poverty_rate, pollution)
fit = slope * poverty_rate + intercept
plt.plot(poverty_rate, pollution, linewidth=0, marker='o')
plt.plot(poverty_rate, fit, 'r--')

plt.grid(True)
plt.savefig("PR_vs_PM2.52014.png")

plt.show()
```

```python
file_df[["City", "PM2.5 2014", "Economic Rank"]].head()
```

```python
pollution = file_df["PM2.5 2014"]
economic_rank  = file_df["Economic Rank"]
plt.scatter(economic_rank, pollution, marker="o", facecolors="black", edgecolors="blue",
            alpha=0.75)

plt.legend(loc="best", fontsize="small", fancybox=True)
plt.title("Economic Rank vs. PM2.5 2014")

plt.xlabel("Economic Rank by City")
plt.ylabel("PM2.5 2014 g/m3 (micrograms per cubic meter)")

(slope, intercept, r_value, p_value, std_err) = stats.linregress(economic_rank, pollution)
fit = slope * economic_rank + intercept
plt.plot(economic_rank, pollution, linewidth=0, marker='o')
plt.plot(economic_rank, fit, 'r--')

plt.grid(True)
plt.savefig("ER_vs_PM2.52014.png")

plt.show()
```

```python
print ("r-squared:", r_value**2)
```
