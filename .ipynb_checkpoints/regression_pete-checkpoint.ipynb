```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress
```

```python
raw_data = "resources/AQ_health_econ.csv"

state_FIPS = "resources/state-fips.csv"

AQ_health_econ_df = pd.read_csv(raw_data)

state_FIPS_df = pd.read_csv(state_FIPS)

state_FIPS_df.dtypes
```

```python
AQ_health_econ_df.head(20)
```

```python
# Takes last five digits of FIPS and creates"FIPS_place", to match census FIPS for places
AQ_health_econ_df['FIPS'] = AQ_health_econ_df['FIPS'].astype(str)

AQ_health_econ_df['FIPS_place'] = AQ_health_econ_df['FIPS'].str[-5:]
```

```python
AQ_health_econ_df.head()
```

```python
AQ_health_econ_df['State'] = AQ_health_econ_df['State'].astype(str)

state_FIPS_df['State'] = state_FIPS_df['State'].astype(str)

AQ_health_econ_df['State'] = AQ_health_econ_df['State'].str.strip()

state_FIPS_df['State'] = state_FIPS_df['State'].str.strip()

merged_df = pd.merge(AQ_health_econ_df,state_FIPS_df,on='State', how='inner')

merged_df.head(140)
```

```python
plt.figure(figsize=(20,16))
plt.scatter(AQ_health_econ_df['PM2.5 2014'], AQ_health_econ_df['Pulminary Disease %'], color='b', alpha=0.75)
plt.title("Pulminary Disease Prevalence Vs. PM2.5 Pollution")
plt.xlabel("2014 PM2.5 Pollution (micrograms per cubic meter)")
plt.ylabel("Prevalence of Pulminary Disease")
```

```python
plt.figure(figsize=(20,16))
plt.scatter(AQ_health_econ_df['PM2.5 2014'], AQ_health_econ_df['Asthma %'], color='b', alpha=0.75)
plt.title("Ashtma % Vs. PM2.5 Pollution")
plt.xlabel("2014 PM2.5 Pollution (micrograms per cubic meter)")
plt.ylabel("Prevalence of Asthma")
```

```python
x_axis = AQ_health_econ_df['PM2.5 2014']
y_axis = AQ_health_econ_df['Pulminary Disease %']

(slope, intercept, _, _, _) = linregress(x_axis, y_axis)
fit = slope * x_axis + intercept

fig, ax = plt.subplots()

fig.suptitle("Linear Regression: PM2.5 Pollution vs. Pulminary Disease", fontsize=16, fontweight="bold")


ax.set_xlabel("2014 PM2.5 Pollution (micrograms per cubic meter)", fontsize=16)
ax.set_ylabel("Prevalence of Pulminary Disease", fontsize=16)

ax.plot(x_axis, y_axis, linewidth=0, marker='o', label =AQ_health_econ_df['City'])
ax.plot(x_axis, fit, 'b--')

#for City in Cities:

#plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             #arrowprops=dict(facecolor='black', shrink=0.05),
            # )

plt.gcf().set_size_inches(10, 8)

plt.show()
```
