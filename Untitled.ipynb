```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
file = "Resources/AQ_health_econ.csv"
file_df = pd.read_csv(file)
file_df
```

```python
file_df[["City", "PM2.5 2014", "Poverty Rate 2009-2013"]].head()
```

```python
pollution = file_df["PM2.5 2014"]
poverty_rate  = file_df["Poverty Rate 2009-2013"]

plt.scatter(pollution, poverty_rate, marker="o", facecolors="red", edgecolors="black",
            alpha=0.75)

plt.legend(loc="best", fontsize="small", fancybox=True)
plt.title("PM2.5 2014 vs. Poverty Rate")

plt.xlabel("PM2.5 2014")
plt.ylabel("Poverty Rate 2009-2013")

plt.grid(True)
plt.savefig("PM2.52014_vs_PR.png")

plt.show()
```

```python
file_df[["City", "PM2.5 2014", "Economic Rank"]].head()
```

```python
pollution = file_df["PM2.5 2014"]
economic_rank  = file_df["Economic Rank"]
plt.scatter(pollution, economic_rank, marker="o", facecolors="blue", edgecolors="black",
            alpha=0.75)

plt.legend(loc="best", fontsize="small", fancybox=True)
plt.title("PM2.5 2014 vs. Economic Rank")

plt.xlabel("PM2.5 2014")
plt.ylabel("Economic Rank")

plt.grid(True)
plt.savefig("PM2.52014_vs_ER.png")

plt.show()
```
