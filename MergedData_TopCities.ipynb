```python
# Import Dependencies
import pandas as pd
```

```python
citylist_csv = ("Resources/City_List.csv")
life_expectancy_csv = ("Resources/Life Expectancy_2015.csv")
```

```python
cili_df = pd.read_csv(citylist_csv)
liex_df = pd.read_csv(life_expectancy_csv) 
```

```python
cili_df.head()
```

```python
liex_df.head()
```

```python
li_ex_df = liex_df.iloc[:, [0, 1, 3]]
```

```python
li_ex_df.head()
```

```python
merge_table = pd.merge(cili_df, li_ex_df,  how='left', left_on=['State','County'], right_on = ['State','County'])
```

```python
merge_table.head(100)
```

```python
average_le = merge_table.groupby('City').mean()
average_le.head(100)
```

```python
new_table = pd.merge(merge_table, average_le, on="City", how='inner')
new_table
```

```python
new_table1 = new_table.drop_duplicates(subset='City', keep='first')
print(new_table1)
```

```python
new_table1.iloc[:, [0, 1, 5, 4, 3, 6]]
new_table1.rename(columns = {'State_x':'State', 'FIPS_x':'FIPS', 
                              'Life Expectancy_x':'Life Expectancy'}, inplace = True)
new_table1
```

```python
new_table1.to_excel("Output/City_List_Updated.xlsx", index=False)
new_table1.to_csv("Output/City_List_Updated.csv")
```
