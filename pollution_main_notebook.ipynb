```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
raw_data = "resources/AQ_and_health.csv"

AQ_health_df = pd.read_csv(raw_data)

AQ_health_df.dtypes
```

```python
plt.figure(figsize=(10,8))
plt.scatter(AQ_health_df['Pulminary Disease %'], AQ_health_df['PM2.5 2014'], color='b', alpha=0.75)
plt.title("Pulminary Disease Prevalence Vs. PM2.5 Pollution")
plt.xlabel("Prevalence of Pulminary Disease")
plt.ylabel("2014 PM2.5 Pollution (micrograms per cubic meter)")
```

```python
plt.figure(figsize=(10,8))
plt.scatter(AQ_health_df['Asthma %'], AQ_health_df['PM2.5 2014'], color='b', alpha=0.75)
plt.title("Asthma Prevalence Vs. PM2.5 Pollution")
plt.xlabel("Prevalence of Asthma")
plt.ylabel("2014 PM2.5 Pollution (micrograms per cubic meter)")
```

```python
econ_data = "resources/econ_data.csv"

econ_df = pd.read_csv(econ_data)

econ_df.dtypes
```

```python
AQ_health_econ_df = pd.merge(AQ_health_df,econ_df[['FIPS','Poverty Rate 2009-2013','Economic Rank']],on='FIPS', how='inner')

AQ_health_econ_df.head(20)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(AQ_health_econ_df['Pulminary Disease %'], AQ_health_econ_df['Poverty Rate 2009-2013'], color='b', alpha=0.75)
plt.title("Pulminary Disease Prevalence Vs. Poverty")
plt.xlabel("Prevalence of Pulminary Disease")
plt.ylabel("Percent Poverty 2009-2013")
```
