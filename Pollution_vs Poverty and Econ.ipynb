```python
#Import dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

```python
#Read csv file into notebook
file = "Resources/AQ_health_econ.csv"
file_df = pd.read_csv(file)
file_df
```

```python
#Show City, Pollution, and Economic Rank from csv
file_df[["City", "PM2.5 2014", "Economic Rank"]].head()
```

```python
#Pulling data from Pollution and Ecomonic Rank Columns
pollution = file_df["PM2.5 2014"]
economic_rank  = file_df["Economic Rank"]

#Formatting scatter plot of data
plt.scatter(pollution, economic_rank, marker="o", facecolors="black", edgecolors="blue",
            alpha=0.75)

plt.title("PM2.5 2014 vs. Economic Rank", fontsize=16)
plt.xlabel("PM2.5 2014 g/m3 (micrograms per cubic meter)", fontsize=16)
plt.ylabel("Economic Rank by City", fontsize=16)

#Insert code for regression line
(slope, intercept, r_value, p_value, std_err) = stats.linregress(pollution, economic_rank)
fit = slope * pollution + intercept
plt.plot(pollution, economic_rank, linewidth=0, marker='o')
plt.plot(pollution, fit, 'b--')

plt.grid(True)
plt.savefig("PM2.52014_vs_ER.png")
plt.gcf().set_size_inches(10, 10)

#Show Economic Rank of City vs. Pollution of 72 Cities in the US
plt.show()
```

```python
#Show City, Pollution, and Economic Rank from csv
file_df[["City", "PM2.5 2014", "Poverty Rate 2009-2013"]].head()
```

```python
#Pulling data from Pollution and Poverty Rate Columns
pollution = file_df["PM2.5 2014"]
poverty_rate = file_df["Poverty Rate 2009-2013"]
plt.scatter(pollution, poverty_rate, marker="o", facecolors="blue", edgecolors="black",
            alpha=0.75)

#Formatting scatter plot of data
plt.title("PM2.5 2014 vs. Poverty Rate 2009-2013", fontsize=16)
plt.xlabel("PM2.5 2014 g/m3 (micrograms per cubic meter)", fontsize=16)
plt.ylabel("Poverty Rate 2009-2013", fontsize=16)

#Insert code for regression line
(slope, intercept, r_value, p_value, std_err) = stats.linregress(pollution, poverty_rate)
fit = slope * pollution + intercept
plt.plot(pollution, poverty_rate, linewidth=0, marker='o')
plt.plot(pollution, fit, 'b--')

plt.grid(True)
plt.savefig("PM2.52014_vs_PD.png")
plt.gcf().set_size_inches(10, 10)

#Show Poverty Rate vs. Pollution of 72 Cities in the US
plt.show()
```

```python
#Print R Squared Value - Roughly 4% due to the outliers of the data. Shows how the dependent variable
#(poverty rate) is explained by the independent variable (pollution).
print ("r-squared:", r_value**2)
```
