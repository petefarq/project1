```python
import pandas as pd
import numpy as np
import gmaps
import requests
from config import api_key
api_key = api_key
```

```python
data_file = "output_data\AQ_and_health.csv"
health_df = pd.read_csv(data_file)
health_df = health_df.dropna(how="any", inplace = False)
health_df
```

```python
gmaps.configure(api_key=api_key)
```

```python
locations = health_df[["Lat", "Lon"]].astype(float)
particulate = health_df["PM2.5 2014"].astype(float)
```

```python
fig = gmaps.figure(layout={'width': '500px',
                          'height': '500px',
                          'padding': '3px',
                          'border': '1px solid black'})
heat_layer = gmaps.heatmap_layer(locations, weights = particulate, 
                                dissipating = False, max_intensity = 30,
                                point_radius = 1.5)
fig.add_layer(heat_layer)

fig
```

```python
asthma = health_df['Asthma %'].astype(float)
fig2 = gmaps.figure(layout={'width': '500px',
                          'height': '500px',
                          'padding': '3px',
                          'border': '1px solid black'})

heat_layer2 = gmaps.heatmap_layer(locations, weights = asthma, 
                                dissipating = False, max_intensity = 30,
                                point_radius = 1.5)

fig2.add_layer(heat_layer2)

fig2
```

```python
co2 = health_df['CO2 2012'].astype(float)

fig3 = gmaps.figure(layout={'width': '500px',
                          'height': '500px',
                          'padding': '3px',
                          'border': '1px solid black'})

heat_layer3 = gmaps.heatmap_layer(locations, weights = co2, 
                                dissipating = False, max_intensity = .9,
                                point_radius = 2)
fig3.add_layer(heat_layer3)

fig3
```
