```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
raw_data = "resources/top_100_population.csv"

top_100_df = pd.read_csv(raw_data)

top_100_df.dtypes
```

```python
top_100_df.head(10)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(top_100_df['Population 2015'], top_100_df['PM2.5 2014'], color='b', alpha=0.75)
plt.title("Population Vs. PM2.5 Pollution")
plt.xlabel("2015 Population, 10xmillion")
plt.ylabel("2014 PM2.5 Pollution")
```
