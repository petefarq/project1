```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#import CSV file
health_csv = ("Resources/AQ_and_health.csv")
all_data_csv = ("Output/City_List_Updated.csv")
```

```python
#create dataframe
health_df = pd.read_csv(health_csv)
all_data_df = pd.read_csv(all_data_csv) 
```

```python
#show dataframe
health_df
```

```python
#show dataframe
all_data_df
```

```python
#merge above dataframes
health_all_data = pd.merge(all_data_df, health_df, on="City", how='outer')
health_all_data
```

```python
#change column names
health_all_data = health_all_data.rename(columns=
                                        {'State_x': 'State',
                                        'Cancer %': 'Cancer Percentage',
                                        'Heart Disease %': 'Heart Disease Percentage',
                                        'PM2.5 2014': 'Pollution 2014'})
health_all_data
```

```python
# Show plot
health_all_data.plot.scatter(x='Pollution 2014', y='Cancer Percentage', figsize=(8,6))
#add title
plt.title("Pollution in 2014 vs. Cancer %")
#save plot
plt.savefig("Images/2014_pollution_vs._Cancer%.png")
plt.show()
```

```python
# Show plot
health_all_data.plot.scatter(x='Pollution 2014', y='Life Expectancy', figsize=(8,6))
#change title
plt.title("Pollution in 2014 vs. Life Expectancy")
#save figure to .png
plt.savefig("Images/2014_pollution_vs._Life_Expectancy.png")
plt.show()
```

```python
# Show plot
health_all_data.plot.scatter(x='Pollution 2014', y='Heart Disease Percentage', figsize=(8,6))
#change title
plt.title("Pollution in 2014 vs. Heart Disease %")
#save figure to .png
plt.savefig("Images/2014_pollution_vs._Heart_Disease%.png")
plt.show()
```
